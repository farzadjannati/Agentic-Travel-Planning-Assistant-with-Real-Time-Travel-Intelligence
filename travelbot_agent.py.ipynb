در این بخش لازم است با استفاده از ابزار مناسب مجموعه داده (zinovadr/iata-airport-code) را از بستر Kaggle دانلود نموده و یک Dictionary mapping هوشمند از نام شهر به کد فرودگاه مورد نظر بسازید.

<p dir='rtl' style="line-height: 2.0; text-align: right; font-family: Vazir; font-size: 16px; margin-top: 20px; color: white; background-color:rgb(0, 40, 30); padding: 30px; border-radius: 8px;">
🎯 <b>خروجی مورد انتظار:</b><br>
• تعداد فرودگاه‌های دارای کد IATA<br>
• تعداد ورودی‌های mapping dictionary<br>
• نمونه جستجو: Dubai → DXB، Tehran → THR

````python
# WRITE YOUR CODE HERE
# Download IATA airport code dataset from Kaggle
import kagglehub

# Download the dataset
iata_path = kagglehub.dataset_download("zinovadr/iata-airport-code")
print(f"Dataset downloaded to: {iata_path}")

# Find and load the CSV file
iata_files = os.listdir(iata_path)
print(f"Files in dataset: {iata_files}")

# Load the dataset
iata_df = None
for f in iata_files:
    if f.endswith('.csv'):
        iata_df = pd.read_csv(os.path.join(iata_path, f))
        break

print(f"Dataset shape: {iata_df.shape}")
print(f"Columns: {list(iata_df.columns)}")
iata_df.head()
````

````python
# WRITE YOUR CODE HERE
# Build smart city-to-IATA mapping dictionary
# Clean and process the dataframe
iata_df_clean = iata_df.dropna(subset=['iata_code']).copy()
iata_df_clean = iata_df_clean[iata_df_clean['iata_code'].str.strip() != '']

# Build mapping: city name (lowercase) -> IATA code
city_to_iata = {}
for _, row in iata_df_clean.iterrows():
    city = str(row.get('municipality', '')).strip().lower()
    iata_code = str(row['iata_code']).strip().upper()
    
    if city and iata_code and len(iata_code) == 3:
        # If city already in dict, prefer larger/main airports
        if city not in city_to_iata:
            city_to_iata[city] = iata_code

# Also add name field as alternative lookup
for _, row in iata_df_clean.iterrows():
    name = str(row.get('name', '')).strip().lower()
    iata_code = str(row['iata_code']).strip().upper()
    if name and iata_code and len(iata_code) == 3:
        if name not in city_to_iata:
            city_to_iata[name] = iata_code

# Print statistics
print(f"Number of airports with IATA code: {len(iata_df_clean)}")
print(f"Number of entries in mapping dictionary: {len(city_to_iata)}")

# Sample lookups
test_cities = ['dubai', 'tehran', 'paris', 'london', 'tokyo', 'istanbul', 'new york']
print("\nSample lookups:")
for city in test_cities:
    code = city_to_iata.get(city, 'NOT FOUND')
    print(f"  {city.title()} → {code}")


def get_iata_code(city_name: str) -> str:
    """Smart IATA code lookup with fuzzy matching."""
    city_lower = city_name.strip().lower()
    
    # Direct lookup
    if city_lower in city_to_iata:
        return city_to_iata[city_lower]
    
    # Partial match
    for key, value in city_to_iata.items():
        if city_lower in key or key in city_lower:
            return value
    
    # If input is already a 3-letter IATA code
    if len(city_name.strip()) == 3 and city_name.strip().isalpha():
        return city_name.strip().upper()
    
    return None


# Test the smart lookup function
print("\nSmart lookup tests:")
for city in ['Dubai', 'Tehran', 'New York', 'Los Angeles', 'DXB']:
    result = get_iata_code(city)
    print(f"  {city} → {result}")
````

````python
# WRITE YOUR CODE HERE
# Download currency and country code mapping dataset from Kaggle
currency_path = kagglehub.dataset_download("phanee16/currency-and-country-code-mapping")
print(f"Dataset downloaded to: {currency_path}")

# Find and load the CSV file
currency_files = os.listdir(currency_path)
print(f"Files in dataset: {currency_files}")

currency_df = None
for f in currency_files:
    if f.endswith('.csv'):
        currency_df = pd.read_csv(os.path.join(currency_path, f))
        break

print(f"Dataset shape: {currency_df.shape}")
print(f"Columns: {list(currency_df.columns)}")
currency_df.head()
````

````python
# WRITE YOUR CODE HERE
# Build smart country-to-currency mapping dictionary
country_to_currency = {}

for _, row in currency_df.iterrows():
    country = str(row.get('Country', row.get('country', ''))).strip().lower()
    currency_code = str(row.get('Code', row.get('code', row.get('AlphabeticCode', row.get('Currency Code', ''))))).strip().upper()
    
    if country and currency_code and len(currency_code) == 3 and currency_code != 'NAN':
        country_to_currency[country] = currency_code

# Add common aliases
common_aliases = {
    'usa': 'USD', 'us': 'USD', 'united states': 'USD', 'united states of america': 'USD', 'america': 'USD',
    'uk': 'GBP', 'united kingdom': 'GBP', 'england': 'GBP', 'britain': 'GBP', 'great britain': 'GBP',
    'uae': 'AED', 'united arab emirates': 'AED', 'emirates': 'AED',
    'iran': 'IRR', 'japan': 'JPY', 'china': 'CNY', 'india': 'INR',
    'france': 'EUR', 'germany': 'EUR', 'italy': 'EUR', 'spain': 'EUR',
    'turkey': 'TRY', 'russia': 'RUB', 'brazil': 'BRL', 'canada': 'CAD',
    'australia': 'AUD', 'south korea': 'KRW', 'korea': 'KRW',
    'saudi arabia': 'SAR', 'qatar': 'QAR', 'kuwait': 'KWD',
    'thailand': 'THB', 'malaysia': 'MYR', 'singapore': 'SGD',
    'egypt': 'EGP', 'mexico': 'MXN', 'switzerland': 'CHF',
}
for alias, code in common_aliases.items():
    if alias not in country_to_currency:
        country_to_currency[alias] = code

# Print statistics
unique_currencies = set(country_to_currency.values())
print(f"Number of unique currencies: {len(unique_currencies)}")
print(f"Number of entries in mapping dictionary: {len(country_to_currency)}")

# Sample lookups
test_countries = ['iran', 'uae', 'united states', 'france', 'japan', 'turkey']
print("\nSample lookups:")
for country in test_countries:
    code = country_to_currency.get(country, 'NOT FOUND')
    print(f"  {country.upper()} → {code}")


def get_currency_code(country_name: str) -> str:
    """Smart currency code lookup with fuzzy matching."""
    country_lower = country_name.strip().lower()
    
    # Direct lookup
    if country_lower in country_to_currency:
        return country_to_currency[country_lower]
    
    # Partial match
    for key, value in country_to_currency.items():
        if country_lower in key or key in country_lower:
            return value
    
    # If input is already a 3-letter currency code
    if len(country_name.strip()) == 3 and country_name.strip().isalpha():
        return country_name.strip().upper()
    
    return None


# Test smart lookup
print("\nSmart lookup tests:")
for c in ['Iran', 'UAE', 'USA', 'France', 'EUR']:
    print(f"  {c} → {get_currency_code(c)}")
````

````python
# WRITE YOUR CODE HERE
# Initialize the embedding model
import logging
logging.getLogger("sentence_transformers").setLevel(logging.ERROR)

embedding_model = SentenceTransformer("BAAI/bge-small-en-v1.5")
EMBEDDING_DIM = 384  # Dimension of bge-small-en-v1.5

print(f"Embedding model loaded: BAAI/bge-small-en-v1.5")
print(f"Embedding dimension: {EMBEDDING_DIM}")


# Define LanceDB schemas using Pydantic
class FAQSchema(LanceModel):
    question: str
    answer: str
    vector: Vector(EMBEDDING_DIM)


class BookSchema(LanceModel):
    page_number: int
    text: str
    vector: Vector(EMBEDDING_DIM)


# Connect to LanceDB
db = lancedb.connect("./travelbot_lancedb")
print("Connected to LanceDB")

# =============================================
# Prepare FAQ embeddings and store in LanceDB
# =============================================
faq_records = []
for item in faq_data:
    question = item.get('question', item.get('Q', item.get('q', '')))
    answer = item.get('answer', item.get('A', item.get('a', '')))
    
    # Create embedding from the question text for semantic search
    combined_text = f"{question} {answer}"
    embedding = embedding_model.encode(combined_text).tolist()
    
    faq_records.append({
        'question': question,
        'answer': answer,
        'vector': embedding
    })

# Create FAQ table
faq_table = db.create_table("faq", data=faq_records, mode="overwrite")
print(f"FAQ table created with {len(faq_records)} entries")

# =============================================
# Prepare Book embeddings and store in LanceDB
# =============================================
book_records_lance = []
for _, row in book_df.iterrows():
    text = row['text']
    
    # Split long texts into chunks for better retrieval
    chunk_size = 1000
    overlap = 200
    
    if len(text) > chunk_size:
        chunks = []
        start = 0
        while start < len(text):
            end = min(start + chunk_size, len(text))
            chunks.append(text[start:end])
            start += chunk_size - overlap
    else:
        chunks = [text]
    
    for chunk in chunks:
        if chunk.strip():
            embedding = embedding_model.encode(chunk).tolist()
            book_records_lance.append({
                'page_number': int(row['page_number']),
                'text': chunk,
                'vector': embedding
            })

# Create Book table
book_table = db.create_table("travel_book", data=book_records_lance, mode="overwrite")
print(f"Travel Book table created with {len(book_records_lance)} entries")

# =============================================
# Test search in both tables
# =============================================
print("\n--- Test FAQ Search ---")
test_query = "What documents do I need for travel?"
query_embedding = embedding_model.encode(test_query).tolist()
faq_results = faq_table.search(query_embedding).limit(2).to_pandas()
for _, row in faq_results.iterrows():
    print(f"Q: {row['question'][:100]}")
    print(f"A: {row['answer'][:100]}")
    print()

print("\n--- Test Book Search ---")
test_query_book = "best tourist attractions in Paris"
query_embedding_book = embedding_model.encode(test_query_book).tolist()
book_results = book_table.search(query_embedding_book).limit(2).to_pandas()
for _, row in book_results.iterrows():
    print(f"Page {row['page_number']}: {row['text'][:150]}...")
    print()

print("\nVector database built and tested successfully!")
````

<div dir="rtl" style="text-align: right; padding:10px; background-color:#6B7280;  border-radius: 12px; border: 2px solid rgb(2, 34, 22); font-family: Vazir;">
یکی از ابزارهای مهمی که تراول‌بات با استفاده از آن به کاربران خدمات می‌دهد، ابزار جستجوی پرواز است. شما باید ابزاری تعریف کنید که پارامترهایی مانند مبدا، مقصد و تاریخ پرواز را از کاربر دریافت کرده و با استفاده از Amadeus API به جستجوی پروازهای مورد نظر بپردازد. این ابزار باید حداقل ۳ پرواز را با اطلاعات کاملی شامل شماره پرواز، کد ایرلاین، قیمت دقیق (به همراه واحد پول)، تاریخ و ساعت حرکت، مدت پرواز و تعداد توقف‌ها به کاربر بازگرداند.

**نکات مهم پیاده‌سازی:**

- **تبدیل نام شهر به کد IATA:** ابزار Amadeus بر اساس کدهای یکتای فرودگاه (IATA Code) کار می‌کند. از تابع مپینگی که در بخش قبل نوشته‌اید استفاده کنید تا نام شهرها را به کد فرودگاه تبدیل کنید (مثلاً Tehran → IKA، Dubai → DXB).

- **مدیریت فرمت تاریخ:** Amadeus تاریخ را به فرمت YYYY-MM-DD می‌پذیرد. از آنجایی که ماهیت چت‌بات صحبت به زبان طبیعی است، کاربران فرمت تاریخ را به گونه‌های متفاوتی وارد می‌کنند(برای مثال به صورت متن و نه عدد). شما باید مدیریت فرمت تاریخ در ابزار و System Prompt را مد نظر داشته باشید.

- **مدیریت خطا:** در صورت بروز خطا در جستجو (مثل عدم یافتن پرواز یا خطای API)، پیام خطای مناسب و قابل فهم به کاربر بازگردانید.

- **استفاده از Decorator:** از `@tool` decorator برای تعریف این ابزار استفاده کنید تا LangGraph بتواند آن را شناسایی کند.

</div>


````python
# WRITE YOUR CODE HERE
@tool
def search_flights(origin: str, destination: str, departure_date: str) -> str:
    """Search for flights between two cities using Amadeus API.
    
    Args:
        origin: Origin city name or IATA code (e.g., 'Tehran', 'Dubai', 'IKA')
        destination: Destination city name or IATA code (e.g., 'Paris', 'CDG')
        departure_date: Departure date in YYYY-MM-DD format (e.g., '2025-12-20')
    
    Returns:
        Formatted string with flight search results including flight number,
        airline, price, departure time, duration, and stops.
    """
    try:
        # Convert city names to IATA codes
        origin_code = get_iata_code(origin)
        destination_code = get_iata_code(destination)
        
        if not origin_code:
            return f"Error: Could not find IATA code for origin city '{origin}'. Please provide a valid city name or IATA code."
        if not destination_code:
            return f"Error: Could not find IATA code for destination city '{destination}'. Please provide a valid city name or IATA code."
        
        # Validate and parse date
        try:
            parsed_date = date_parser.parse(departure_date)
            formatted_date = parsed_date.strftime("%Y-%m-%d")
        except Exception:
            return f"Error: Invalid date format '{departure_date}'. Please use YYYY-MM-DD format (e.g., 2025-12-20)."
        
        # Search flights using Amadeus API
        response = amadeus.shopping.flight_offers_search.get(
            originLocationCode=origin_code,
            destinationLocationCode=destination_code,
            departureDate=formatted_date,
            adults=1,
            max=5
        )
        
        flights = response.data
        
        if not flights:
            return f"No flights found from {origin} ({origin_code}) to {destination} ({destination_code}) on {formatted_date}."
        
        # Format the results
        result = f"✈️ Flight Results: {origin} ({origin_code}) → {destination} ({destination_code}) on {formatted_date}\n"
        result += "=" * 70 + "\n\n"
        
        for i, flight in enumerate(flights[:5], 1):
            price = flight['price']['total']
            currency = flight['price']['currency']
            
            for segment in flight['itineraries'][0]['segments']:
                carrier = segment['carrierCode']
                flight_number = f"{carrier}{segment['number']}"
                departure_time = segment['departure']['at']
                arrival_time = segment['arrival']['at']
                duration = flight['itineraries'][0]['duration']
                stops = len(flight['itineraries'][0]['segments']) - 1
            
            result += f"Flight {i}:\n"
            result += f"  Flight Number: {flight_number}\n"
            result += f"  Airline Code: {carrier}\n"
            result += f"  Price: {price} {currency}\n"
            result += f"  Departure: {departure_time}\n"
            result += f"  Arrival: {arrival_time}\n"
            result += f"  Duration: {duration.replace('PT', '').lower()}\n"
            result += f"  Stops: {stops}\n\n"
        
        return result
    
    except ResponseError as error:
        return f"Amadeus API Error: {error.response.body}"
    except Exception as e:
        return f"Error searching flights: {str(e)}"


# Test the flight search tool
print(search_flights.invoke({"origin": "Tehran", "destination": "Dubai", "departure_date": "2026-01-14"}))
````

- **مدیریت خطا:** در صورت عدم یافتن هتل یا خطای API، پیام مناسب به کاربر بازگردانید.

- **Fallback به Tavily:** در صورتی که Amadeus Hotel API در دسترس نبود یا محدودیت داشت، می‌توانید به عنوان fallback از Tavily Search استفاده کنید.

- **استفاده از Decorator:** از `@tool` decorator برای تعریف این ابزار استفاده کنید.
<div>


````python
# WRITE YOUR CODE HERE
@tool
def search_hotels(destination: str, check_in_date: str, check_out_date: str, budget: str = "") -> str:
    """Search for hotels in a destination city using Amadeus API with Tavily fallback.
    
    Args:
        destination: Destination city name or IATA code (e.g., 'Dubai', 'Paris')
        check_in_date: Check-in date in YYYY-MM-DD format (e.g., '2025-12-20')
        check_out_date: Check-out date in YYYY-MM-DD format (e.g., '2025-12-25')
        budget: Optional budget range (e.g., 'low', 'medium', 'high')
    
    Returns:
        Formatted string with hotel search results.
    """
    try:
        # Convert city name to IATA code
        city_code = get_iata_code(destination)
        if not city_code:
            return f"Error: Could not find IATA code for '{destination}'."
        
        # Parse dates
        try:
            parsed_checkin = date_parser.parse(check_in_date).strftime("%Y-%m-%d")
            parsed_checkout = date_parser.parse(check_out_date).strftime("%Y-%m-%d")
        except Exception:
            return f"Error: Invalid date format. Please use YYYY-MM-DD format."
        
        try:
            # Step 1: Search for hotels by city
            hotel_list_response = amadeus.reference_data.locations.hotels.by_city.get(
                cityCode=city_code
            )
            
            hotel_ids = [h['hotelId'] for h in hotel_list_response.data[:10]]
            
            if not hotel_ids:
                raise Exception("No hotels found via Amadeus")
            
            # Step 2: Get hotel offers
            hotel_offers_response = amadeus.shopping.hotel_offers_search.get(
                hotelIds=hotel_ids,
                checkInDate=parsed_checkin,
                checkOutDate=parsed_checkout,
                adults=1
            )
            
            offers = hotel_offers_response.data
            
            if not offers:
                raise Exception("No hotel offers found")
            
            result = f"🏨 Hotel Results in {destination} ({city_code})\n"
            result += f"Check-in: {parsed_checkin} | Check-out: {parsed_checkout}\n"
            result += "=" * 70 + "\n\n"
            
            for i, hotel in enumerate(offers[:5], 1):
                hotel_name = hotel['hotel'].get('name', 'N/A')
                hotel_rating = hotel['hotel'].get('rating', 'N/A')
                lat = hotel['hotel'].get('latitude', 'N/A')
                lon = hotel['hotel'].get('longitude', 'N/A')
                
                if hotel.get('offers'):
                    offer = hotel['offers'][0]
                    price = offer['price'].get('total', 'N/A')
                    currency = offer['price'].get('currency', 'N/A')
                else:
                    price = 'N/A'
                    currency = ''
                
                result += f"Hotel {i}:\n"
                result += f"  Name: {hotel_name}\n"
                result += f"  Rating: {hotel_rating} stars\n"
                result += f"  Price: {price} {currency} per night\n"
                result += f"  Location: ({lat}, {lon})\n\n"
            
            return result
            
        except Exception as amadeus_error:
            # Fallback to Tavily Search
            print(f"Amadeus hotel search failed, falling back to Tavily: {amadeus_error}")
            
            query = f"best hotels in {destination} prices rating {parsed_checkin} to {parsed_checkout}"
            tavily_response = tavily_client.search(query=query, max_results=5)
            
            result = f"🏨 Hotel Recommendations for {destination}\n"
            result += f"Check-in: {parsed_checkin} | Check-out: {parsed_checkout}\n"
            result += "(Source: Web Search)\n"
            result += "=" * 70 + "\n\n"
            
            for i, item in enumerate(tavily_response.get('results', []), 1):
                result += f"Hotel {i}:\n"
                result += f"  Title: {item.get('title', 'N/A')}\n"
                result += f"  Description: {item.get('content', 'N/A')[:200]}\n"
                result += f"  URL: {item.get('url', 'N/A')}\n\n"
            
            return result
    
    except Exception as e:
        return f"Error searching hotels: {str(e)}"


# Test hotel search
print(search_hotels.invoke({
    "destination": "Dubai",
    "check_in_date": "2025-03-15",
    "check_out_date": "2025-03-20"
}))
````

  - نواحی مناسب اقامت
  - (در صورت وجود interests) موضوعات خاص کاربر



- **مدیریت اطلاعات ناکافی:** اگر اطلاعات کافی در پایگاه دانش یافت نشد، از Tavily Search استفاده کنید.

- **مدیریت خطا:** خطاهای احتمالی خطای جستجو در پایگاه داده را مدیریت کنید.

- **استفاده از Decorator:** از `@tool` decorator برای تعریف این ابزار استفاده کنید.

**مثال درخواست کاربر:**
```
"Plan my trip to Paris from December 18th to 22th"
"Create a 4-day itinerary for Dubai with focus on food and culture"
```


</div>

````python
## WRITE YOUR CODE HERE
@tool
def plan_trip(destination: str, arrival_date: str, departure_date: str, interests: str = "") -> str:
    """Plan a complete trip itinerary using RAG from the travel book knowledge base.
    
    Args:
        destination: Destination city or country name (e.g., 'Paris', 'Dubai')
        arrival_date: Arrival date in YYYY-MM-DD format (e.g., '2025-12-18')
        departure_date: Departure date in YYYY-MM-DD format (e.g., '2025-12-22')
        interests: User interests like 'culture, food, nature, adventure, shopping' (optional)
    
    Returns:
        Formatted string with a day-by-day trip itinerary with attractions,
        restaurants, and activities.
    """
    try:
        # Parse dates and calculate number of days
        try:
            arrival = date_parser.parse(arrival_date)
            departure = date_parser.parse(departure_date)
            num_days = (departure - arrival).days
            if num_days <= 0:
                num_days = 1
        except Exception:
            return "Error: Invalid date format. Please use YYYY-MM-DD."
        
        # Collect information from travel book using multiple queries
        queries = [
            f"top attractions tourist places in {destination}",
            f"best restaurants food cuisine in {destination}",
            f"activities entertainment things to do in {destination}",
            f"neighborhoods areas to stay in {destination}",
        ]
        
        # Add interest-specific queries
        if interests:
            for interest in interests.split(","):
                queries.append(f"{interest.strip()} in {destination}")
        
        all_context = []
        for q in queries:
            try:
                q_embedding = embedding_model.encode(q).tolist()
                results = book_table.search(q_embedding).limit(3).to_pandas()
                for _, row in results.iterrows():
                    text = row['text']
                    if text not in all_context:
                        all_context.append(text)
            except Exception:
                pass
        
        # If not enough info from book, use Tavily as fallback
        if len(all_context) < 3:
            try:
                tavily_query = f"travel guide {destination} top attractions restaurants activities itinerary"
                tavily_response = tavily_client.search(query=tavily_query, max_results=5)
                for item in tavily_response.get('results', []):
                    all_context.append(item.get('content', ''))
            except Exception:
                pass
        
        # Combine all collected context
        combined_context = "\n\n".join(all_context[:10])
        
        # Build the output
        output = f"🗺️ Trip Plan: {destination}\n"
        output += f"📅 {arrival.strftime('%B %d, %Y')} → {departure.strftime('%B %d, %Y')} ({num_days} days)\n"
        if interests:
            output += f"🎯 Interests: {interests}\n"
        output += "=" * 70 + "\n\n"
        
        output += "📖 Destination Information from Travel Guide:\n"
        output += "-" * 40 + "\n"
        output += combined_context[:2000] + "\n\n"
        
        output += f"\n📋 Suggested {num_days}-Day Itinerary:\n"
        output += "-" * 40 + "\n"
        
        for day in range(1, num_days + 1):
            current_date = arrival + timedelta(days=day - 1)
            output += f"\n🗓️ Day {day} ({current_date.strftime('%B %d, %Y')}):\n"
            output += f"  Morning: Explore attractions and sightseeing\n"
            output += f"  Afternoon: Cultural activities and local experiences\n"
            output += f"  Evening: Dinner at local restaurants and nightlife\n"
        
        output += "\n💡 Note: This itinerary is based on our travel knowledge base. "
        output += "Use the other tools (restaurant search, weather info) for detailed planning.\n"
        
        return output
    
    except Exception as e:
        return f"Error planning trip: {str(e)}"


# Test trip planning
print(plan_trip.invoke({
    "destination": "Paris",
    "arrival_date": "2025-12-18",
    "departure_date": "2025-12-22",
    "interests": "culture, food"
}))
````

````python
# WRITE YOUR CODE HERE
# Define all tools list
tools = [
    search_flights,
    search_hotels,
    search_restaurants,
    get_weather_info,
    get_currency_info,
    search_faq,
    plan_trip
]

# Define comprehensive system message
SYSTEM_MESSAGE = """You are TravelBot, an intelligent travel planning assistant for an online travel agency. You help users with flight bookings, hotel reservations, restaurant recommendations, weather information, currency exchange, answering frequently asked questions, and creating trip itineraries.

You have access to the following 7 tools:

1. **search_flights**: Search for flights between cities. Requires origin city, destination city, and departure date in YYYY-MM-DD format.
2. **search_hotels**: Search for hotels in a city. Requires destination city, check-in date, and check-out date in YYYY-MM-DD format. Optional: budget level.
3. **search_restaurants**: Find top restaurants and local cuisine in a destination. Requires only the destination city name.
4. **get_weather_info**: Get weather forecast for a destination. Requires destination city and date in YYYY-MM-DD format.
5. **get_currency_info**: Get exchange rate between two countries' currencies. Requires origin country name and destination country name.
6. **search_faq**: Search FAQ knowledge base for common travel questions. Requires the user's query text.
7. **plan_trip**: Create a trip itinerary. Requires destination, arrival date, and departure date in YYYY-MM-DD format. Optional: user interests.

IMPORTANT INSTRUCTIONS:
- When users mention dates in natural language (e.g., "December 20th", "next Monday", "March fifteenth 2025"), you MUST convert them to YYYY-MM-DD format before passing to any tool. Today's date is {today}.
- When users mention city names, pass them directly to the tools - the tools handle city-to-IATA conversion internally.
- When users mention country names for currency exchange, pass the country names directly - the tool handles country-to-currency conversion.
- Always be helpful, informative, and provide clear, well-formatted responses.
- Respond ONLY in English.
- If you're unsure about a date or parameter, ask the user for clarification.
- For trip planning, try to gather destination, dates, and interests from the user.
""".format(today=datetime.now().strftime("%Y-%m-%d"))

# Bind tools to the model
model_with_tools = model.bind_tools(tools)

# Verify tools binding
print(f"Successfully bound {len(tools)} tools to the model:")
for t in tools:
    print(f"  - {t.name}")
````

## <div style="text-align: center; direction: rtl; font-family: Vazir;">Creating Graph Nodes</div>


<div dir="rtl" style="text-align: right; padding:10px; background-color:#6B7280;  border-radius: 12px; border: 2px solid rgb(2, 34, 22); font-family: Vazir;">
<p style="line-height: 1.8; text-align: right;">
Reasoner node را پیاده‌سازی کنید که:
- System message و تاریخچه مکالمه را ترکیب کند
- LLM را invoke کند
- پاسخ را برگرداند
</p>
</div>


<p dir='rtl' style="line-height: 2.0; text-align: right; font-family: Vazir; font-size: 16px; margin-top: 20px; color: white; background-color:rgb(0, 40, 30); padding: 30px; border-radius: 8px;">
🎯 <b>خروجی مورد انتظار:</b><br>
تعریف موفق reasoner node
</p>

````python
# WRITE YOUR CODE HERE
# Define the reasoner node
def reasoner(state: MessagesState):
    """Reasoner node that combines system message with conversation history and invokes the LLM."""
    sys_msg = SystemMessage(content=SYSTEM_MESSAGE)
    # Combine system message with all conversation messages
    messages = [sys_msg] + state["messages"]
    response = model_with_tools.invoke(messages)
    return {"messages": [response]}

print("Reasoner node defined successfully!")
````

## <div style="text-align: center; direction: rtl; font-family: Vazir;">Creating and Compiling Graph</div>


<div dir="rtl" style="text-align: right; padding:10px; background-color:#6B7280;  border-radius: 12px; border: 2px solid rgb(2, 34, 22); font-family: Vazir;">
<p style="line-height: 1.8; text-align: right;">
StateGraph را بسازید و compile کنید و تصویر گراف را رسم کنید.

</p>
</div>


<p dir='rtl' style="line-height: 2.0; text-align: right; font-family: Vazir; font-size: 16px; margin-top: 20px; color: white; background-color:rgb(0, 40, 30); padding: 30px; border-radius: 8px;">
🎯 <b>خروجی مورد انتظار:</b><br>
• Graph compiled موفق<br>
• تصویر visualization گراف<br>
</p>

````python
# WRITE YOUR CODE HERE
# Create the StateGraph
graph = StateGraph(MessagesState)

# Add nodes
graph.add_node("reasoner", reasoner)
graph.add_node("tools", ToolNode(tools))

# Add edges
graph.add_edge(START, "reasoner")

# Add conditional edge: if the model calls a tool, go to tools node; otherwise, end
from langgraph.graph import END

def should_continue(state: MessagesState):
    """Determine whether to continue to tools or end."""
    last_message = state["messages"][-1]
    if hasattr(last_message, "tool_calls") and last_message.tool_calls:
        return "tools"
    return END

graph.add_conditional_edges("reasoner", should_continue, {"tools": "tools", END: END})
graph.add_edge("tools", "reasoner")

# Compile the graph
travel_agent = graph.compile()
print("Graph compiled successfully!")

# Visualize the graph
try:
    graph_image = travel_agent.get_graph().draw_mermaid_png()
    display(Image(graph_image))
    print("Graph visualization displayed above.")
except Exception as e:
    print(f"Could not render graph image: {e}")
    # Fallback: print the mermaid diagram
    print(travel_agent.get_graph().draw_mermaid())
````

````python
# WRITE YOUR CODE HERE
def chat_cli():
    """Interactive CLI chat interface for TravelBot."""
    print("=" * 60)
    print("🌍 Welcome to TravelBot - Your Intelligent Travel Assistant!")
    print("=" * 60)
    print("Type your travel questions below.")
    print("Commands: 'exit' to quit, 'clear' to reset conversation")
    print("-" * 60)
    
    conversation_history = []
    
    while True:
        try:
            user_input = input("\n🧑 You: ").strip()
            
            if not user_input:
                continue
            
            if user_input.lower() == 'exit':
                print("\n👋 Thank you for using TravelBot! Have a great trip!")
                break
            
            if user_input.lower() == 'clear':
                conversation_history = []
                print("\n🔄 Conversation cleared!")
                continue
            
            # Add user message to history
            conversation_history.append(HumanMessage(content=user_input))
            
            # Invoke the graph
            response = travel_agent.invoke({"messages": conversation_history})
            
            # Extract the final response and tools used
            tools_used = []
            for msg in response["messages"]:
                if hasattr(msg, "tool_calls") and msg.tool_calls:
                    for tc in msg.tool_calls:
                        tools_used.append(tc["name"])
            
            # Get the final AI message
            final_message = response["messages"][-1]
            
            # Display tools used
            if tools_used:
                print(f"\n🔧 Tools used: {', '.join(tools_used)}")
            
            # Display the response
            print(f"\n🤖 TravelBot: {final_message.content}")
            
            # Update conversation history with full response
            conversation_history = response["messages"]
            
        except KeyboardInterrupt:
            print("\n\n👋 Goodbye!")
            break
        except Exception as e:
            print(f"\n❌ Error: {str(e)}")
            continue


# Uncomment to run CLI (comment out for testing section)
# chat_cli()
print("CLI function defined. Uncomment 'chat_cli()' to start interactive chat.")
````

<h2 style="color: #24292e; border-bottom: 1px solid #eaecef; padding-bottom: 8px; margin-top: 30px;">
Conclusion & System Assessment
</h2>

### 1. Strengths
*   **Robust Entity Resolution:** The system effectively maps natural language entities (cities, countries) to technical codes (IATA, ISO 4217), which is crucial for API interaction.
*   **Resilient Architecture:** The implementation of **Fallbacks** (e.g., trying Tavily if Amadeus fails) ensures the user rarely sees a "I don't know" error.
*   **Effective RAG:** The usage of LanceDB with `BAAI/bge-small-en-v1.5` allows the bot to answer specific policy questions and provide rich travel content without hallucinating.
*   **Clean Flow:** LangGraph manages the state transitions effectively, passing context between the Reasoner and Tools nodes.

### 2. Weaknesses
*   **API Rate Limits:** Heavy reliance on free-tier APIs (Amadeus/Tavily) can lead to occasional failures or limited result sets (e.g., only 2-3 hotels).
*   **Context Window:** While the RAG retrieves info, compressing a whole chapter of a travel book into a short chat response involves information loss.
*   **Date Parsing:** While handled well, extreme edge cases in date formats might still confuse the `dateutil` parser.

### 3. Suggestions for Improvement
*   **User Memory:** Implement a persistent database (like SQLite or Postgres) for LangGraph Checkpointing to remember user preferences across different sessions.
*   **Interactive Refinement:** Add a "human-in-the-loop" node to ask clarifying questions if the budget or specific interests are ambiguous before calling tools.
*   **Multi-Modal Output:** Enhance the CLI or UI to display images of hotels/destinations returned by the search tools.
